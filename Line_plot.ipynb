# Line plot

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# (1, 300)은 300차원 짜리 데이터 1개 -> (300, 1)
y = np.random.randn(300, 1)
y.shape
```

```python
fig, ax = plt.subplots(figsize=(10, 7))

ax.plot(y, label='noise', color='skyblue')
ax.axhline(y=0, color='r', linewidth=3)

# tick setting
x_ticks = np.arange(0, 301, 100)
ax.set_xticks(x_ticks)
ax.tick_params(labelsize=20)

# label setting
ax.set_xlabel("time(t)", fontsize=20)
ax.set_ylabel("Amplitude(amp)", fontsize=20)

# legend setting
ax.legend(fontsize=20, bbox_to_anchor=(1, 0.7), loc='center left')
```

```python
import cv2
from google.colab.patches import cv2_imshow
def show_img(imgfile, fx=1, fy=1):
    imgfile =  imgfile
    img=cv2.imread(imgfile, cv2.IMREAD_COLOR)
    crop_img_file = cv2.resize(img, dsize=(0, 0), fx=fx, fy=fy, interpolation=cv2.INTER_LINEAR)
    cv2_imshow(crop_img_file)
```

```python
# data
name_list = ['filter/natural test', 'filter/adversarial test',
             'column/natural test', 'column/adversarial test',
             'irregular/natural test', 'irregular/adversarial test']

data_dict = {'FN':[50, 65, 70, 75, 76], 
             'FA':[34, 38, 41, 42, 43], 
             'CN':[61, 68, 70, 74, 76], 
             'CA':[36, 40, 42, 43, 43], 
             'IN':[76, 78, 77, 78, 76], 
             'IA':[42, 41, 44, 44, 43]
             }

plt.rcParams['lines.linewidth'] = 2

# line plot setting
color_list = ['r', 'r', 'g', 'g', 'c', 'c']
marker_list = ['v', 'o', '^', 'd', 's', '*']
linestyle_list = ['-', '--', '-', '--', '-', '--']

fig, ax = plt.subplots(figsize=(9, 9))

for idx, line in enumerate(data_dict.values()):
    ax.plot(line, 
            linestyle=linestyle_list[idx], 
            color=color_list[idx], 
            alpha=0.8,
            marker=marker_list[idx], 
            markersize=12,
            label=name_list[idx])

ax.set_yticks(np.linspace(0, 100, 6))
ax.tick_params(labelsize=15)

ax.grid(axis='x', linewidth=0)
ax.grid(axis='y', linewidth=0)

ax.set_xlabel("w", fontsize=20)
ax.set_ylabel("Accuracy(%)", fontsize=20)

ax.legend(ncol=2, loc='lower center', fontsize=15)
```
